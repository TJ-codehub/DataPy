#### **Grundregeln:**

| **In Jupyter** | **In Streamlit** |
|----------------|------------------|
| `print(...)` | `st.write(...)` |
| `df.head()` | `st.dataframe(df)` |
| `plt.show()` | `st.pyplot(fig)` |
| Zellen einzeln ausführen | Gesamte Datei läuft neu bei Interaktion |
| Variablen bleiben im Speicher | `st.cache_data` oder `st.cache_resource` |

#### **`app.py` - Hauptseite (Beispiel)**

```python
import streamlit as st
import pandas as pd

st.set_page_config(page_title="Projektname", page_icon="📊", layout="wide")

st.title("📊 [Ihr Projekttitel]")
st.markdown(""" Diese App analysiert [dein Thema] und erstellt Vorhersagen.
👈 **Wähle eine Seite im Sidebar!** """)

@st.cache_data
def load_data():
    return pd.read_csv('data/cleaned_data.csv')
df = load_data()

# Quick Stats
col1, col2, col3 = st.columns(3)
col1.metric("Datensätze", len(df))
col2.metric("Features", len(df.columns))
col3.metric("[Ihre eigene Metrik]", df['target'].sum())  # z.B. Anzahl kranke Menschen
st.dataframe(df.head())
```

#### **`pages/1_Daten_Exploration.py` - Übersicht & Filter**

```python
import streamlit as st
import pandas as pd

st.title("Daten Exploration")

@st.cache_data
def load_data():
    return pd.read_csv('data/cleaned_data.csv')
df = load_data()

# Sidebar: Filter
st.sidebar.header("Filter Optionen")
age_col = 'age'  # hier: Ihre eigene Spalte
if age_col in df.columns:
    age_range = st.sidebar.slider(
        "Alter:",
        int(df[age_col].min()),
        int(df[age_col].max()),
        (int(df[age_col].min()), int(df[age_col].max()))
    )
    df = df[(df[age_col] >= age_range[0]) & (df[age_col] <= age_range[1])]

# Tabs
tab1, tab2, tab3 = st.tabs(["📊 Übersicht", "📈 Statistiken", "🔢 Rohdaten"])
with tab1:
    col1, col2 = st.columns(2)
    col1.metric("Gefilterte Zeilen", len(df))
    col2.metric("Spalten", len(df.columns))
with tab2:
    st.dataframe(df.describe())
with tab3:
    st.dataframe(df)
```

## Häufige Fehler

| **Fehler** | **Lösung** |
|------------|------------|
| `NameError: name 'st' is not defined` | `import streamlit as st` vergessen |
| Plot wird nicht angezeigt | `plt.show()` → `st.pyplot(fig)` |
| App ist langsam | `@st.cache_data` für Datenlade-Funktion |
| Pfad nicht gefunden | `'data/file.csv'` relativ zum Projekt-Root |
| Modell wird jedes Mal neu trainiert | `st.session_state` verwenden |
