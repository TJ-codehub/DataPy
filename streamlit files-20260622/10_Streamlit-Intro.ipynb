## The Problem

# Why Streamlit?

## Advantages

| Feature | Description |
|---------|-------------|
| 🚀 **Fast** | App in minutes, not days |
| 🐍 **Python only** | No new languages to learn |
| 📊 **Interactive** | Widgets, sliders, filters |
| 🔄 **Live updates** | Change code → app updates instantly |
| 🌐 **Shareable** | Host & share for free |

## Perfect for

- ✅ Project presentations
- ✅ Interactive dashboards
- ✅ ML model demos
- ✅ Data explorers

# Streamlit vs. Jupyter Notebook

## Differences

| Aspect | Jupyter Notebook | Streamlit |
|--------|-----------------|----------|
| **Purpose** | Development & exploration | Presentation & interaction |
| **Code** | Visible | Hidden |
| **Interactivity** | Static | Dynamic |
| **Audience** | Data scientists | Everyone (incl. non-coders) |
| **Output** | Notebook file | Web app (URL) |

## Workflow

```
1. 📓 Analyse in Jupyter Notebook
2. 🧹 Clean up the code
3. 🎨 Convert to Streamlit app
4. 🚀 Present & share!
```

# First Interactivity: Slider

```python
import streamlit as st
import pandas as pd

st.title('Interactive Filter 🎛️')

# Load data
df = pd.read_csv('heart_disease_cleaned.csv')

# Age slider
age_filter = st.slider(
    'Select age:',
    min_value=int(df['age'].min()),
    max_value=int(df['age'].max()),
    value=(30, 70)  # default range
)

# Filter data
filtered_df = df[
    (df['age'] >= age_filter[0]) & 
    (df['age'] <= age_filter[1])
]

# Show result
st.write(f'Showing **{len(filtered_df)}** of {len(df)} patients')
st.write(f'Age between {age_filter[0]} and {age_filter[1]} years')
st.dataframe(filtered_df)
```

# Caching for Performance

## Problem: The script re-runs on EVERY interaction!

```python
# WITHOUT caching – SLOW! ❌
def load_data():
    df = pd.read_csv('large_file.csv')  # Runs on every slider change!
    return df

df = load_data()  # Called over and over again
age = st.slider('Age:', 0, 100, 50)
```

## Solution: `@st.cache_data`

```python
# WITH caching – FAST! ✅
@st.cache_data  # Load once, then served from cache!
def load_data():
    df = pd.read_csv('large_file.csv')
    return df

df = load_data()  # Only loaded the first time
age = st.slider('Age:', 0, 100, 50)  # Fast!
```

**Important:** Use caching for:
- Loading data
- Expensive computations
- API calls

# App Structure: Best Practices

```python
import streamlit as st
import pandas as pd

# 1. Page config (ALWAYS FIRST!)
st.set_page_config(
    page_title='Heart Disease Dashboard',
    page_icon='🫀',
    layout='wide',  # or 'centered'
    initial_sidebar_state='expanded'
)

# 2. Caching for data
@st.cache_data
def load_data():
    return pd.read_csv('heart_disease_cleaned.csv')

# 3. Main function
def main():
    st.title('Heart Disease Dashboard 🫀')
    
    # Load data
    df = load_data()
    
    # Sidebar
    st.sidebar.title('Filters')
    age_range = st.sidebar.slider('Age:', 20, 80, (30, 70))
    
    # Main area
    st.write(f'{len(df)} patients')
    st.dataframe(df)

# 4. Run
if __name__ == '__main__':
    main()
```

# Complete Example: Dashboard

**File: `app.py`**

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# Page config
st.set_page_config(page_title='Heart Disease Dashboard', page_icon='🫀')

@st.cache_data
def load_data():
    return pd.read_csv('heart_disease_cleaned.csv')

def main():
    # Title
    st.title('Heart Disease Dashboard 🫀')
    st.write('Interactive Data Analysis')
    
    # Load data
    df = load_data()
    
    # Sidebar: filters
    st.sidebar.title('Filters ⚙️')
    age_range = st.sidebar.slider(
        'Age:',
        int(df['age'].min()),
        int(df['age'].max()),
        (30, 70)
    )
    
    show_condition = st.sidebar.radio(
        'Show:',
        ['All', 'Diseased only', 'Healthy only']
    )
    
    # Filter data
    filtered_df = df[
        (df['age'] >= age_range[0]) & 
        (df['age'] <= age_range[1])
    ]
    
    if show_condition == 'Diseased only':
        filtered_df = filtered_df[filtered_df['condition'] == 1]
    elif show_condition == 'Healthy only':
        filtered_df = filtered_df[filtered_df['condition'] == 0]
    
    # Main area
    st.subheader('Filtered Data')
    st.write(f'Showing **{len(filtered_df)}** of {len(df)} patients')
    
    # Metrics
    col1, col2, col3 = st.columns(3)
    col1.metric('Patients', len(filtered_df))
    col2.metric('Avg. Age', f"{filtered_df['age'].mean():.1f}")
    col3.metric('Avg. Cholesterol', f"{filtered_df['chol'].mean():.0f}")
    
    # Show data
    if st.checkbox('Show raw data'):
        st.dataframe(filtered_df)
    
    # Visualisation
    st.subheader('Visualisation')
    feature = st.selectbox(
        'Select feature:',
        ['age', 'chol', 'trestbps', 'thalach']
    )
    
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.hist(filtered_df[feature], bins=20, edgecolor='black', color='skyblue')
    ax.set_xlabel(feature)
    ax.set_ylabel('Frequency')
    ax.set_title(f'Distribution: {feature}')
    ax.grid(True, alpha=0.3)
    st.pyplot(fig)
    
    # Footer
    st.markdown('---')
    st.write('Data source: Heart Disease Cleveland UCI')

if __name__ == '__main__':
    main()
```

# From Jupyter to Streamlit

## Step-by-step conversion

**1. Jupyter Notebook code:**
```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
print(f"Rows: {len(df)}")

plt.hist(df['age'])
plt.show()
```

**2. Streamlit app code:**
```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

@st.cache_data
def load_data():
    return pd.read_csv('data.csv')

df = load_data()
st.write(f"Rows: {len(df)}")  # print → st.write

fig, ax = plt.subplots()
ax.hist(df['age'])
st.pyplot(fig)  # plt.show() → st.pyplot()
```

## Key changes

- `print()` → `st.write()`
- `plt.show()` → `st.pyplot(fig)`
- Add caching
- Add interactive elements
