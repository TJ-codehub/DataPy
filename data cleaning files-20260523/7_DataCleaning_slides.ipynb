# 2. Missing Data

## Identification

```python
# Identify missing values
df.isnull().sum()
df.isna().sum()

# Visualize
import missingno as msno
msno.matrix(df)
```

## Three Main Approaches

| Method | Function | When? |
|---------|----------|-------|
| Delete | `dropna()` | Many missing values |
| Replace | `fillna()` | Single missing values |
| Interpolate | `interpolate()` | Time series |

# Strategies for Missing Values

## 1. Remove rows/columns
- When many (or non-critical) values are missing
- `df.dropna()`

## 2. Replace with constant
- Mean, median, mode, or 0
- `df.fillna(df.mean())`

## 3. Forward/Backward Fill
- Time series: use previous/next value
- `df.fillna(method='ffill')`

## 4. Interpolation
- Estimate values between known points
- `df.interpolate()`

# 3. Detecting & Removing Duplicates

```python
# Find duplicates
df.duplicated().sum()

# Show duplicates
df[df.duplicated()]

# Remove duplicates
df.drop_duplicates()

# Check specific columns only
df.drop_duplicates(subset=['name'])
```

**Note:** Parameters `keep='first'`, `'last'` or `False`

# 4. Outliers

## Detection Methods
- IQR Method (Interquartile Range)
- Z-Score
- Visualization (Boxplot) --> very important!

## Handling Outliers
- Remove
- Winsorizing
- Transformation

```python
# IQR Method
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

# 5. Data Type Conversion

```python
# Check data types
df.dtypes

# Convert
df['age'] = df['age'].astype(int)
df['date'] = pd.to_datetime(df['date'])
df['category'] = df['category'].astype('category')

# Multiple columns at once
df = df.astype({
    'col1': 'int64',
    'col2': 'float64'
})
```

# 6. Consistency & Standardization

```python
# Clean strings
df['name'] = df['name'].str.strip()
df['name'] = df['name'].str.lower()
df['name'] = df['name'].str.title()

# Standardize categories
df['sex'].replace({
    'M': 'male', 'm': 'male',
    'F': 'female', 'f': 'female'
}, inplace=True)
```

**Goal:** Uniform spellings and formats!
