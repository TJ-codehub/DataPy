```python
class Person():
    def __init__(self, alter):
        self.alter = alter
        
    def check_age(self):
        if self.alter > 18:
            return True
        else:
            return False

class Student(Person):
    def __init__(self, alter, studiengang):
        if type(alter) is not int:
            raise ValueError("Alter muss int sein")
        if type(studiengang) is not str:
            raise ValueError("studiengang muss str sein")

        self.alter = alter
        self.studiengang = studiengang
        
    def __str__(self):
        return f'Student mit Alter {self.alter} und Studiengang {self.studiengang}.'

student = Student(19, 'Medieninformatik')
student.__dict__
```
